```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_bias(np_arr):
   return np.abs(np.mean(np_arr) - np.median(np_arr))/(np.max(np_arr)-np.min(np_arr))*100


def import_excel(path):
    df = pd.read_excel(path, skiprows=3)
    df = df.dropna(axis=0, how='any')
    df = df.drop(index = [1,3])
    return df


def get_mode(np_arr):
    
    vals, counts = np.unique(np_arr, return_counts= True )
    if np.max(counts) == 1:
        return np.nan
    return np.argwhere(counts == np.max (counts))


def print_desc_stats(np_arr):
    print(f'Мода - {get_mode(np_arr)}')
    print(f'Среднее - {np.mean(np_arr)}')
    print(f'Медиана - {np.median(np_arr)}')
    print(f'Смещение медианы относительно среднего - {get_bias(np_arr)}%')
```
